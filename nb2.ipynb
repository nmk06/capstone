```python
import librosa
import librosa.display
from scipy.io import wavfile as wav
import IPython.display as ipd
import numpy as np
import os
import matplotlib.pyplot as plt

cwd = os.getcwd()
print (cwd)

# Random File
pathToSample = cwd + '//UrbanSound8K//audio//fold5//100852-0-0-0.wav'
librosa_audio, librosa_sample_rate = librosa.load(pathToSample)
scipy_sample_rate, scipy_audio = wav.read(pathToSample)

# Widget Display
data2, sample_rate2 = librosa.load(pathToSample) # 22.05 kHz
librosa.display.waveplot(data2, sr=sample_rate2)
test = ipd.Audio(pathToSample)
display(test)

# Compare the sample rates
print('Different in sample rates, SCIPY and LIBROSA: ', scipy_sample_rate, librosa_sample_rate)
```

```python
# Bit Depth

print('Original audio file range', np.min(scipy_audio), 'to' , np.max(scipy_audio))
print('Librosa audio file ranges', np.min(librosa_audio), 'to' , np.max(librosa_audio))
```

```python
#Original audio with 2 channels
## Ours is two channels
plt.figure(figsize=(10,5))
print('SciPy Audio displayed below')
plt.plot(scipy_audio)

plt.figure(figsize=(10,5))
print('Librosa Audio displayed below')
plt.plot(librosa_audio)
```

```python
# Extract Features

# MFCC displays audio on the mel scale for human auditory system. 40 is the number of coefficiencts to extract. The output calculates
# 40 MFCCs over 173 frames (ie. samples) - 40 rows, ,173 columns
mfccs = librosa.feature.mfcc(y = librosa_audio, sr = librosa_sample_rate, n_mfcc=40)
print(mfccs)
```

```python
# Display spectrogram

import librosa.display
librosa.display.specshow(mfccs, sr=librosa_sample_rate,x_axis = 'time')
```
