```python
# Data Metadata - Importing
metadata = pd.read_csv(cwd + "//UrbanSound8K//metadata//UrbanSound8K.csv")
listOfCategories = metadata.class_name.unique() # Unique Categories of Classes

# Class Distributions - inequality is concern
print(metadata.class_name.value_counts())

# TEMPORARY MODIFICATION
#metadata = metadata.head(5)
#display(metadata)
```

```python
# Audio Sample File Properties

# Load various imports 
import pandas as pd
import os
import librosa
import librosa.display
import soundfile as sf

#from wavfilehelper import WavFileHelper
audiodata = []

# Temporarily SLICE relevant categories
categories = ['dog_bark', 'car_horn', 'gun_shot', 'siren']

for index, row in metadata.iterrows():
    
    # Create data array per wav file
    data = []
    
    # Extract filename and category
    file_str = row["slice_file_name"]
    category_str = row["class_name"]
    
    if category_str in categories:
        data.append(file_str)
        data.append(category_str)

        # Extracting data from each wav file
        file_name_path = os.path.join(os.path.abspath(cwd + "//UrbanSound8K//audio//"),'fold'+str(row["fold"])+'//',str(row["slice_file_name"]))
        
        # Folder
        fold = row["fold"]
        data.append(fold)
        
        # Extract relevant properties
        wave_file = sf.SoundFile(file_name_path)

        num_channels = wave_file.channels
        data.append(num_channels)
        sample_rate = wave_file.samplerate
        data.append(sample_rate)
        sub_type = wave_file.subtype #Bit rate depending on whether it's PCM or ADPCM

        # temp
        #data.append(sub_type)

        # Extract bit depth from sub_type string
        # Documentation cited FLOAT is 32 bit, DOUBLEis 64 bit
        # ADPCM is 4 bit and non existent in categories of choice
        # Collective datatypes in all 8k -> ['PCM_16' 'PCM_24' 'PCM_U8' 'FLOAT' 'IMA_ADPCM' 'MS_ADPCM']
        sub_type = sub_type.replace('PCM_16', '16')
        sub_type = sub_type.replace('PCM_24', '24')
        sub_type = sub_type.replace('PCM_U8', '8')
        sub_type = sub_type.replace('FLOAT', '32')
        
        data.append(sub_type)

        # Append relevant information to audiodata file
        audiodata.append(data)

    else:
        continue

# Convert into a Panda dataframe - MAIN FRAME OF INTEREST
audioDF = pd.DataFrame(audiodata, columns=['wave_file_name','sound_category','folder','num_channels','sample_rate','bit_depth'])

#print(audioDF.sound_category.unique()) PASSS YAY
```
